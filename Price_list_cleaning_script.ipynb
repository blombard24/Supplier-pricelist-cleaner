```python
import pandas as pd
```

```python
df = pd.read_csv('Damasio Drywall - Day to Day - Mar 1st 2022.csv')
```

```python
df
```

```python
# Lots of completely blank rows - Drop Empty Rows
df = df.dropna(axis = 0, how = 'all')
```

```python
df
```

```python
# drop columns not required for future use

drop_columns = ['Page #1', 'Damasio Drywall']
df = df.drop(columns = drop_columns)
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df
```

```python
df.columns
```

```python
df.rename(columns = {'CUSTOMER # ':'Material','Unnamed: 3':'Price'}, inplace = True)
```

```python
# Strip out leading and trailing whitespace in column 'Material'
df['Material'] = df['Material'].str.strip()
```

```python
# Strip out leading and trailing whitespace in column 'Price'
df['Price'] = df['Price'].str.strip()
```

```python
# Drop rows with empty string in columns Material and Price
df.drop(index = (df[df['Material'] == ""].index), inplace = True)
df.drop(index = (df[df['Price'] == ""].index), inplace = True)
# Drop rows that have "date" as a value
df.drop(index = (df[df['Material'].str.contains('date', case = False, na = False)].index), inplace = True)
# Drop remaining NAN rows
df.drop(index = (df[(df['Material'].isna()) | (df['Price'].isna())].index), inplace = True)
```

```python
# Cleaning looks complete, last thing to do is reset the index if required for future use
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
#output to csv to be used in current price book

df.to_csv('updated_cleaned_price_list.csv', index = False)
```
